# Passos 

Importar pandas e numpy


```python
import pandas as pd
import numpy as np
```

index_col = None significa que não possui coluna de índice

```python
receitas_candidatos = pd.read_excel('receitascandidatos1.xls', index_col=None, na_values=['NA'])
```

Seleção dos dados desejados


```python
print(receitas_candidatos.head())
print(receitas_candidatos.shape)
```

```python
df = receitas_candidatos[['UF', 'Nome candidato', 'Tipo receita', 'Valor receita']]
```

```python
print(df.head())
print(df.shape)
```

Filtrar por SP, RJ, MG, DF, BA, e RN

```python
uf_filter = df[(df.UF.isin(['SP', 'RJ', 'MG', 'DF', 'BA', 'RN']))]
```

```python
print(uf_filter.head())
print(uf_filter.shape)
```

Reordenar dados usando pivot table

```python
uf_pivot_table = pd.pivot_table(
    uf_filter, 
    index=['UF', 'Nome candidato'], 
    columns=['Tipo receita'], 
    values='Valor receita', 
    aggfunc=np.sum, 
    fill_value=0
)
```

```python
uf_pivot_table.sort_values(by='Recursos de outros candidatos', ascending=[False])
```
